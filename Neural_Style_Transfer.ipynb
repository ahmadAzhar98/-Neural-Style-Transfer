```python
import tensorflow_hub as hub
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
model = hub.load('https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2')
```

**Image Preprocessing and Load**

```python
def load_image(img_path):
  img = tf.io.read_file(img_path)
  img = tf.image.decode_image(img,channels=3)
  img = tf.image.convert_image_dtype(img,tf.float32)
  img = img[tf.newaxis,:]
  return img
```

```python
content_image = load_image('/content/C_image7.jpg')
style_image = load_image('/content/S_image5.jpg')
```

```python
plt.imshow(np.squeeze(content_image))
plt.show()
```

```python
plt.imshow(np.squeeze(style_image))
plt.show()
```

**Model Training**

```python
trained_image = model(tf.constant(content_image),tf.constant(style_image))[0]
```

```python
plt.imshow(np.squeeze(trained_image))
plt.show()
```
